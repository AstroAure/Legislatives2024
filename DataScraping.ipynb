# Scraping complet

```python
DATA_RESU = []
DATA_STAT = []
dep_list = pd.read_csv("departement_2022.csv")['LIBELLE'].to_list()
for dep in dep_list:
    number = 1
    while True:
        try:
            region, department = scraping.code_circo(dep)
            url = scraping.url_code(region=region, department=department, number=number)
            data_results, data_stats = scraping.get_results_url(url)
            for data in [data_results, data_stats]:
                data.insert(0, 'Departement', dep)
                data.insert(1, 'Circonscription', number)
                data.insert(2, 'Code region', f"{region}")
                data.insert(3, 'Code departement', f"{department}")
                data.insert(4, 'Code circonscription', f"{department}{number:0>2}")
            DATA_RESU.append(data_results)
            DATA_STAT.append(data_stats)
            print(dep, number)
            number += 1
        except:
            break

data_results = pd.concat(DATA_RESU, ignore_index=True)
data_stats = pd.concat(DATA_STAT, ignore_index=True)

print(len(np.unique(data_results['Code circonscription'])))

data_results.to_csv('results_legislatives_2024_T1.csv')
data_stats.to_csv('stats_legislatives_2024_T1.csv')
```
